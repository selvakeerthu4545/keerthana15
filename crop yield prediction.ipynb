```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Sample data for training
years = np.array([2010, 2011, 2012, 2013, 2014, 2015]).reshape(-1, 1)
crop_yield = np.array([9, 11, 13, 15, 17, 19])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(years, crop_yield, test_size=0.2, random_state=42)

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the Mean Squared Error to evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Predict yield for a given year
year_to_predict = np.array([2016]).reshape(-1, 1)
predicted_yield = model.predict(year_to_predict)
print(f"Predicted yield for {year_to_predict[0][0]}: {predicted_yield[0]}")

# Plot the training data and regression line
plt.scatter(years, crop_yield, color='blue')
plt.plot(years, model.predict(years), color='red')
plt.xlabel('Year')
plt.ylabel('Crop Yield')
plt.title('Crop Yield Prediction')
plt.show()
```
